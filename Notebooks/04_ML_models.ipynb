```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import sys
import os
import gensim.downloader

sys.path.append(os.path.abspath("../Code"))

import vectorize
import ml_models

df = pd.read_csv('../Data/tcga_simple_train.csv')
w2v = gensim.downloader.load('word2vec-google-news-300')


X_train, X_test, y_train, y_test = train_test_split(df['text'], df['t'], test_size=0.2, random_state=23)

m_train = vectorize.vectorizacionEmbeddings(X_train,w2v)
m_test = vectorize.vectorizacionEmbeddings(X_test,w2v)

y_pred = ml_models.svm_fit_predict(m_train,m_test,y_train)

print(classification_report(y_test, y_pred))
```
