```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import sys
import os
import gensim.downloader

sys.path.append(os.path.abspath("../Code"))

import vectorize
from ml_models import LogisticModel

df = pd.read_csv('../Data/raw/tcga_simple_train.csv')
w2v = gensim.downloader.load('word2vec-google-news-300')


X_train, X_test, y_train, y_test = train_test_split(df['text'], df['t'], test_size=0.2, random_state=23)

```

```python
m_train, v = vectorize.vectorizacionBinaria(X_train)
m_test = v.transform(X_test)

clf = LogisticModel()
y_pred = clf.fit_predict(m_train,m_test,y_train)
print(classification_report(y_test, y_pred))
```

```python
df2 = pd.read_csv('../Data/raw/tcga_simple_dev_shuffled.csv')

X = v.transform(df2['text'])
y = clf.predict(X)

df2["t"] = y

df2.to_csv('../Results/output.csv', index=False)
```
