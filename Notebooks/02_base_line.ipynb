# Creación de un sistema base

Este pequeño sistema predice la etiqueta que más se repite en el conjunto de entrenamiento.

Importamos el código del script `base_line`:

```python
import sys
import os

sys.path.append(os.path.abspath("../Code"))
import base_line as bs
```

Preparamos el conjunto de datos

```python
import pandas as pd
df = pd.read_csv('../data/tcga_simple_train.csv')

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['t'], test_size=0.2, random_state=42)
```

Predecimos la etiqueta que más se repite en el conjutno de train

```python
y_pred = bs.base_system(X_train, y_train, X_test)
```

Métricas de rendimiento:
- F1 Score

```python
bs.evaluate_f1_score(y_test, y_pred)
```

El rendimiento del sistema es **muy malo** puesto que solo esta prediciendo una clase y no tiene en cuenta el texto y no aprende patrones.
