Y también cargamos los modelos de embeddings usando gensim, y el dataframe

```python
import gensim.downloader
import zipfile
import requests
from gensim.models.fasttext import load_facebook_model
import pandas as pd

if not os.path.exists("../Models/wiki.simple.bin"):
        with requests.get("https://dl.fbaipublicfiles.com/fasttext/vectors-wiki/wiki.simple.zip", stream=True) as r:
            r.raise_for_status()
            with open("../Models/wiki.simple.zip", 'wb') as f:
                for chunk in r.iter_content(chunk_size=8192):
                    f.write(chunk)
        
        with zipfile.ZipFile("../Models/wiki.simple.zip", 'r') as zip_ref:
            zip_ref.extractall("../Models")
        
        os.remove("../Models/wiki.simple.zip")

w2v = gensim.downloader.load('word2vec-google-news-300')
model = load_facebook_model("../Models/wiki.simple.bin")

df = pd.read_csv('../Data/tcga_simple_train.csv')
```
